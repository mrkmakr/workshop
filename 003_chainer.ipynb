まず使うモジュールを`import`します

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# numpy                 数値計算
# matplotlib            図の表示


from PIL import Image # 画像の表示


import os             # ディレクトリの操作など

#-- ここまでは vol2 で使ったもの --


#-- ここから chainer で使うもの --
import chainer
import chainer.functions as F
from chainer import initializers
from chainer import Chain, Variable
import chainer.links as L
from chainer import serializers
```

```python
# directory setting 
data_dir = 'data'
model_dir = 'model'
```

###### 1.モデルを定義する

Alex net は以下のような畳み込み層5層、全結合層3層の計8つの層から構成されるネットワークです。
<img src ='data/Alexnet.png'>Chainer でニューラルネットワークを読み込む際には、まずのモデルクラスの以下のように指定します。

```python
class Alex(chainer.Chain):
    insize = 227
    def __init__(self,dropout_rate = 0.0) :
        super(Alex, self).__init__(
            conv1=L.Convolution2D(None,  96, 11, stride=4),
            conv2=L.Convolution2D(None, 256,  5, pad=2),
            conv3=L.Convolution2D(None, 384,  3, pad=1),
            conv4=L.Convolution2D(None, 384,  3, pad=1),
            conv5=L.Convolution2D(None, 256,  3, pad=1),
            fc6=L.Linear(None, 4096),
            fc7=L.Linear(None, 4096),
            fc8=L.Linear(None, 1000),
        )
        self.dropout_rate = dropout_rate
    
    
    def __call__(self, x, layers):
        ret = {}
        en = layers[-1]
        h = self.conv1(x)
        if 'conv1' in layers:
            ret.update({'conv1':h})
            if en == 'conv1':
                return ret
        h = F.max_pooling_2d(F.local_response_normalization(
            F.relu(h)), 3, stride=2) 
        h = F.dropout(self.conv2(h),ratio = self.dropout_rate)
        if 'conv2' in layers:
            ret.update({'conv2':h})
            if en == 'conv2':
                return ret
        h = F.max_pooling_2d(F.local_response_normalization(
            F.relu(h)), 3, stride=2)
        h = F.dropout(self.conv3(h),ratio = self.dropout_rate)
        if 'conv3' in layers:
            ret.update({'conv3':h})
            if en == 'conv3':
                return ret
        h = F.relu(h)
        h = F.dropout(self.conv4(h),ratio = self.dropout_rate)
        if 'conv4' in layers:
            ret.update({'conv4':h})
            if en == 'conv4':
                return ret
        h = F.relu(h)
        h = F.dropout(self.conv5(h),ratio = self.dropout_rate)
        if 'conv5' in layers:
            ret.update({'conv5':h})
            if en == 'conv5':
                return ret
        h = F.max_pooling_2d(F.relu(h), 3, stride=2)
        h = F.dropout(self.fc6(h),ratio = self.dropout_rate)
        if 'fc6' in layers:
            ret.update({'fc6':h})
            if en == 'fc6':
                return ret
        h = F.relu(h)
        h = F.dropout(self.fc7(h),ratio = self.dropout_rate)
        if 'fc7' in layers:
            ret.update({'fc7':h})
            if en == 'fc7':
                return ret
        h = F.relu(h)
        h = self.fc8(h)
        if 'fc8' in layers:
            ret.update({'fc8':h})
        
        return ret
    
    def predict(self,x):
        h = self.__call__(x,layers ='fc8')['fc8']
        h = F.softmax(h)
        return h.data
    
    def get_layer_names(self):
        return ['conv1','conv2','conv3','conv4','conv5','fc6','fc7','fc8']
    
    def get_whole_units(self,layers):
        ret = {}
        if 'conv1' in layers:
            ret.update({'conv1':[96,55,55]})
        if 'conv2' in layers:
            ret.update({'conv2':[256,27,27]})
        if 'conv3' in layers:
            ret.update({'conv3':[384,13,13]})
        if 'conv4' in layers:
            ret.update({'conv4':[384,13,13]})
        if 'conv5' in layers:
            ret.update({'conv5':[256,13,13]})
        if 'fc6' in layers:
            ret.update({'fc6':[4096]})
        if 'fc7' in layers:
            ret.update({'fc7':[4096]})
        if 'fc8' in layers:
            ret.update({'fc8':[1000]})
        return ret
```

このように `Alex` クラスを作り、 `model変数`に代入します。

```python
model = Alex()
```

Alex net を使った画像認識を行います。

```python
# Resize and show image
image = image.resize([227,227])
image
```

まだ `numpy配列`になっていないので変換します

```python
# Convert PIL to numpy.ndarray
imgs = np.asarray(image)
```

この`imgs変数`を Alex net に入力したいのですが、ニューラルネットの仕様上このままだとエラーが出てしまいますので以下の関数を使います。

```python
def preinput(img):
    """ 
    input image ... np.ndarray  227 * 227 * 3 [RGB] 
    
    return img 1 * 3 * 227 * 227 [BGR] sutable for Alex net
    """
    
    img = img.astype(np.float32)
    
    img = img[:,:,[2,1,0]]
    
    
    img = img.transpose(2,0,1).astype(np.float32)

    img = np.expand_dims(img,axis=0)
    
    return img
```

Alex net に画像を入力する前には必ず、この関数の引数に入力画像を入れて、Alex net に入力できる仕様に変換します

```python
img = preinput(imgs)
```

この `img変数`を Alex net に入力して Alex net の予測値を得ます。

```python
# input image to VGG16
#img = prepare(imgs)
#img = np.expand_dims(img,axis=0)
prediction = model.predict(img)
```

この予測値と予測ラベルを対応させて、Alex net が予測した上位10カテゴリを表示します。下の関数を使います。

```python
def show10results(prediction):
    #load label from outsize 
    categories = np.loadtxt(os.path.join(data_dir,'labels.txt'), str, delimiter="\n")

    # sort to higher score represented label
    result = zip(prediction.reshape((prediction.size,)), categories)
    result = sorted(result, reverse=True)

    # show top 10 results 
    print('top  percentage category')
    for i, (score, label) in enumerate(result[:10]):
        print('{:>3d} {:>6.2f}% {}'.format(i + 1, score * 100, label))
```

```python
show10results(prediction)
```

画像のカテゴリ (waterfowl: 水鳥) にドンピシャではありませんが、識別結果の上位に鳥 (black swan: 黒鳥; red-breasted merganser: ウミアイサ) が来ていることがわかります。

# 3. Chainer の `Variableオブジェクト`と勾配計算

Chainer はニューラルネットワークの計算を効率化するために、`numpy.array型` をベースにした Chainer 特有の`Variableオブジェクト`というものがあります。

ここではまず、`numpy.ndarray型`の変数を定義してからそれを `Variable型`を作成します。

```python
x_data = np.array([5]).astype(np.float32) # chainer supports only np.float32, not np.float64 (this is default)
x_data
```

```python
type(x_data)
```

```python
# Variabelize
x = Variable(x_data)
```

```python
# check type
type(x)
```

Variable オブジェクトから元の `numpy.ndarray型`を取り出す場合には "`.data`"を使います

```python
x.data
```

Variable オブジェクトは `numpy.array型`をベースにしているので、基本的な算術演算をサポートしています。

例えば $ y = x^2 -2x +1$ を計算するには以下のように普通に記述できます

```python
y = x ** 2 - 2 * x + 1
y
```

Variabelオブジェクト (`x`) の計算結果 (`y`) もVariableオブジェクトになっています。

Variableオブジェクトの特徴の優れているところは、結果の値以外に、計算の履歴を保持しているところです。

具体的には、yの微分 $\  \frac{dy}{dx} = 2x -2$ を"`.backward()`"で簡単に行うことができます。

```python
y.backward()
```

backward によって計算された勾配が、変数 `x` の`gradフィールド`に保存されます

```python
x.grad
```

`x.grad` の値は $ \frac{dy}{dx} = 2 * 5 -2 = 8$ の$8$を表しています。

以上のことを使って初期値 $5$ から始めて、$y = x^2 -2x + 1$ の最小値とその時の $x$ の値を求めます。

ある $x$ の時の勾配を計算して勾配が小さくなるように $x$ の値を変化させていきます (勾配降下法)

```python
# initial value
xdata = np.array([5]).astype(np.float32)

# Variablize
x = Variable(xdata)
```

```python
# list for saving mid values
x_list = []

# add initial (ndarray) value
x_list.append(x.data[0])     # add float value 

for i in range(30):
    # calcurate y
    y = x ** 2 - 2 * x + 1 
    
    # differential
    y.backward()
    
    # new x ( change based on dy/dx 0.8 is rate
    x.data -=  0.8 * x.grad
    x_list.append(x.data[0])
    
    # remove current gradient 
    x.cleargrad()
```

```python
x_list
```

```python
# to numpy ndarraize 
x_array = np.asarray(x_list)
y_array = x_array ** 2 -2 * x_array +1 
```

```python
# show figure 
fig = plt.figure()
ax = fig.add_subplot(111)
x_ = np.linspace(-4,6,100)
y_ = x_ **2 - 2 * x_ + 1 
ax.plot(x_,y_,color = 'black')
#ax.grid()
ax.plot(x_array,y_array)
ax.scatter(x_array[0],y_array[0],color = 'Orange')
ax.scatter(x_array[-1],y_array[-1],color = 'Orange')
ax.annotate('Start',xy = [x_array[0],y_array[0]],xytext = (x_array[0]-1,y_array[0]+6),
            horizontalalignment = 'center',arrowprops = dict(facecolor='black',arrowstyle = '->' ),
           fontsize = 12)

ax.annotate('End',xy = [x_array[-1],y_array[-1]],xytext = (x_array[-1]-0.7,y_array[-1]-4),
            horizontalalignment = 'center',arrowprops = dict(facecolor='black',arrowstyle = '->'),
           fontsize = 12)
ax.scatter(x_array[0],y_array[0],color = 'Orange')
ax.scatter(x_array[-1],y_array[-1],color = 'Orange')
ax.annotate('$y = x^2 - 2x +1 $',xy = [-3,22],
           fontsize = 12)
ax.set_ylim([0,25])

print('min(y): ' +str(y_array[-1])+', argmin(x): '+str(x_array[-1]))
```

勾配法を用いることで関数の最小値を求めることができました。勾配計算を自動的に行ってくれるところが `Variableオブジェクト`の魅力です。

初期値を違う変数にして同様のことを行ってみます。どの初期値から始めても極小値に落ち着きます。

```python
# initial value (Change 5 to -2)
xdata = np.array([-2]).astype(np.float32)

# Variablize
x = Variable(xdata)


# list for saving mid values
x_list = []

# add initial (ndarray) value
x_list.append(x.data[0])     # add float value 

for i in range(30):
    # calcurate y
    y = x ** 2 - 2 * x + 1 
    
    # differential
    y.backward()
    
    # new x ( change based on dy/dx 0.8 is rate
    x.data -=  0.8 * x.grad
    x_list.append(x.data[0])
    
    # remove current gradient 
    x.cleargrad()
    
# to numpy ndarraize 
x_array = np.asarray(x_list)
y_array = x_array ** 2 -2 * x_array +1 


# show figure 
fig = plt.figure()
ax = fig.add_subplot(111)
x_ = np.linspace(-4,6,100)
y_ = x_ **2 - 2 * x_ + 1 
ax.plot(x_,y_,color = 'black')
#ax.grid()
ax.plot(x_array,y_array)
ax.scatter(x_array[0],y_array[0],color = 'Orange')
ax.scatter(x_array[-1],y_array[-1],color = 'Orange')
ax.annotate('Start',xy = [x_array[0],y_array[0]],xytext = (x_array[0]-1,y_array[0]+6),
            horizontalalignment = 'center',arrowprops = dict(facecolor='black',arrowstyle = '->' ),
           fontsize = 12)

ax.annotate('End',xy = [x_array[-1],y_array[-1]],xytext = (x_array[-1]-0.7,y_array[-1]-4),
            horizontalalignment = 'center',arrowprops = dict(facecolor='black',arrowstyle = '->'),
           fontsize = 12)
ax.scatter(x_array[0],y_array[0],color = 'Orange')
ax.scatter(x_array[-1],y_array[-1],color = 'Orange')
ax.annotate('$y = x^2 - 2x +1 $',xy = [-3,22],
           fontsize = 12)
ax.set_ylim([0,25])

print('min(y): ' +str(y_array[-1])+', argmin(x): '+str(x_array[-1]))
```
